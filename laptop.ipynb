#### Represting relation between attributes of laptop dataset. 

```python
plt.plot(df['brand'])
plt.show()
```

```python
sns.lineplot(df["price"])
plt.show()
```

```python
sns.histplot(df['price'])
plt.show()
```

```python
sns.countplot(x ='warranty',data=df)
plt.title("Warranty of laptops")
plt.xlabel("Years of warranty")
plt.ylabel("Numbers of laptops")
plt.show()
```

```python
counts = df["warranty"].value_counts()
plt.pie(counts,labels=counts.index,autopct='%1.1f%%')
plt.title("Brands of laptop")
plt.show()
##Your pie chart is technically working, but the problem is that one category (91.7%) is much larger than the others, so the small slices (0.7%, 1.0%, 6.6%) overlap and their labels become unreadable.
# If one category is above 80–90%, a bar chart is usually better because pie charts make very small values difficult to see.
```

```python
plt.hist(df['price'], bins=10)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(df['price'], bins=10, kde=True)
plt.title("Price Distribution")
plt.show()
```

```python
plt.scatter(df['Ram'], df['price'])
plt.xlabel("RAM")
plt.ylabel("Price")
plt.title("RAM vs Price")
plt.show()
```

```python
sns.scatterplot(x='Ram', y='price', data=df)
plt.title("RAM vs Price")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='brand', y='price', data=df)
plt.xticks(rotation=90)  # it rotate x labels.
plt.title("Brand vs Price")
plt.show()
```

```python
sns.countplot(x='Ram', data=df)
plt.title("RAM Distribution")
plt.show()
```

```python
sns.countplot(x ="warranty",hue="Ram",data=df)
plt.title("warranty according to ram")
plt.show()
```

```python
avg_price = df.groupby('brand')['price'].mean().sort_values()

avg_price.plot(kind='bar')
plt.title("Average Price by Brand")
plt.ylabel("Price")
plt.show()
```

```python
sns.violinplot(x='brand', y='price', data=df)
plt.xticks(rotation=90)
plt.title("Price Distribution by Company")
plt.show()
```

```python
top10 = df.nlargest(10, 'price')

sns.barplot(x='brand', y='price', data=top10)
plt.title("Top 10 Expensive Laptops")
plt.show()
```

```python
sns.kdeplot(df['display_size'])
plt.show()
```

```python
sns.relplot(x="price",y="display_size",hue = 'ROM_type',data=df)
plt.show()
```

```python
sns.catplot(x = "warranty",y = "price",hue="ROM_type",kind='bar',data=df)
plt.show()
```

```python
ratings = df.groupby("warranty")[[
"spec_rating",
"display_size",
"resolution_width",
"resolution_height"
]].mean()

ratings.plot(kind="bar")
plt.title("Average Service Ratings")
plt.ylabel("Average Rating")
plt.xlabel("Average Rating")
plt.xticks(rotation=0)
plt.show()
```
